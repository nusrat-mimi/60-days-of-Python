```python
ids = (1,2,3)
name = ('nusrat','jahan','mimi')
```

```python
zipped = tuple(zip(ids,name))
```

```python
zipped
```

```python
#unzipped
```

```python
ids_new, name_new = zip(*zipped)
```

```python
ids_new
```

```python
name_new
```

```python
i1, n1 = zipped[0]                     #index wise unzip
```

```python
#Unpacking using loop
```

```python
for ids, name in zipped:
    print(ids,name)
```

```python
#Enumerating: Iterate over pairs

for data in enumerate(zipped):
    #unpack ids, name
    ids, name = data
    print(data)
```

```python
#Enumerating: Iterate over pairs

for index, pair in enumerate(zipped):
    #unpack ids, name
    ids, name = pair
    print('Rank {}, id = {}, Name = {}'.format(index, ids,name))
```
