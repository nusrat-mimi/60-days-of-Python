```python
search_word = 'Barcelona'
replace_word = 'Bayern Munich'

with open (r'text1.txt', 'r') as file:
    data = file.read()
    data = data.replace(search_word, replace_word)
```

```python
with open (r'text1.txt', 'w') as file:
    file.write(data)
```
