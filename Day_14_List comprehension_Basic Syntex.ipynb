# Simple list Compresension
#example:  creating a list of squares for numbers from 0 to 9

```python
squares = [x**2 for x in range(10)]             #x**2 = expression; x=item ; 10= iterable data
print(squares)
```

```python
[x for x in range(10)]         #using list
```

```python
for x in range(10):            #using for loop
    print(x)
```

# *List comprehension with condition*
#creating a list of Even numbers

```python
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

```python
data = 'I love data science'
[x for x in data]
```

```python
data = 'I love data science'
data = data.split()
[x for x in data if x!= 'I']
```

# Creating a matix using list comprehension in python
#using nested list comprehension to create a matrix

```python
matrix = [[j for j in range(5)] for i in range(3)]
print(matrix)
```

# Using functions in list comprehensions
#creating a list of lengths of each word in a list

```python
words = ['Python', 'list', 'Comprehension', 'Example']
lengths = [len(words) for words in words]
print(lengths)
```

```python
words = ['Python', 'list', 'Comprehension', 'Example']
lengths = [len(x) for x in words]
print(lengths)
```
