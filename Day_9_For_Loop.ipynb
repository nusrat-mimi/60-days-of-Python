```python
for i in range(5):
    print('data science')
```

```python
for j in range(5):
    print('data science',j) # this j is loop
```

```python
for i in range(5,10):   #n-1
    print('data science',i)
```

```python
for i in range(5,10,2):   #2=how many times loop continue,increment/decrement, i=i+2
    print('data science',i)
```

```python
for i in range(100,50,-10):
    print('data science',i)
```

```python
for i in range(100,80,-2): #seeing even num between 100 to 80
    print(i)
```

```python
for i in range(100,80,-1): #seeing odd num using conditional
    if i%2 != 0:
        print(i) 
```

```python
for i in range(99,80,-2):
    print(i)  #without conditional
```
