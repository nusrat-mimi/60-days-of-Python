local vs global

```python
x=1000 #global

def func1():
    print ('My Num is: ',x)
func1()

def func2():
    print ('My Num is: ',x)
func2()

def func3():
    x=500 #local
    print ('My Num is: ',x)
func3()
```

```python
x=1000 #global

def func1():
    x=500 #local
    print ('My Num is: ',x)
func1()
print ('My Num is: ',x)
```

```python
def func1():
    global x
    x=500 #local
    print ('My Num is: ',x)
func1()
print ('My Num is: ',x)
```
